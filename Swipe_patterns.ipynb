```python
#LIBRARIES
import matplotlib.pyplot as plt
```

```python
#Function for visualization of patterns.

def visualize_pattern(pattern):
    pos_map = {
        1: (0, 2), 2: (1, 2), 3: (2, 2),
        4: (0, 1), 5: (1, 1), 6: (2, 1),
        7: (0, 0), 8: (1, 0), 9: (2, 0)
    }

    fig, ax = plt.subplots(figsize=(4, 4))
    ax.set_xlim(-1, 3)
    ax.set_ylim(-1, 3)
    ax.set_aspect('equal')
    ax.axis('off')
    fig.set_facecolor('black')

    # Draw dots
    for i in range(1, 10):
        x, y = pos_map[i]
        ax.plot(x, y, 'o', markersize=20, color='black', markeredgecolor=(0, 0.56, 0.06))

    # Draw lines between pattern points
    for i in range(len(pattern) - 1):
        x1, y1 = pos_map[pattern[i]]
        x2, y2 = pos_map[pattern[i + 1]]
        ax.plot([x1, x2], [y1, y2], color=(0,0.6+i/25,0+i/25), linewidth=3)

    # Optionally highlight the sequence
    for idx, p in enumerate(pattern):
        x, y = pos_map[p]
        ax.text(x, y, str(idx + 1), color='black', ha='center', va='center', fontsize=18)
        ax.text(x, y, str(idx + 1), color=(0,0.6+i/25,0+i/25), ha='center', va='center', fontsize=12)

    plt.show()

# Example: pattern connecting 1 → 2 → 5 → 9
visualize_pattern([1, 2, 5, 9 , 6, 3, 4,8,7])
```

```python
visualize_pattern([1, 2, 3, 4,5,6,7,8,9])
possible_directions_simplified = {1:[2,4,5], 2:[1,3,4,5,6],3:[2,5,6],4:[1,2,5,7,8],5:[1,2,3,4,6,7,8,9],6:[2,3,5,8,9],7:[4,5,8],8:[7,4,5,6,9],9:[8,5,6]}
possible_directions_full = {1:[2,4,5,6,8], 2:[1,3,4,5,6,7,8],3:[2,5,6,4,8],4:[1,2,5,7,8,3,9],5:[1,2,3,4,6,7,8,9],6:[2,3,5,8,9,1,7],7:[4,5,8,2,6],8:[7,4,5,6,9,1,3],9:[8,5,6,4,2]}

def genarate_simple_pattern(start=[1],used=[]):
  used = start
  patterns = []
  patterns.append(start)
  for x in possible_directions_simplified[start[-1]]:
    if x not in used:
      patterns.extend(genarate_simple_pattern(start+[x]))
  return patterns
print(genarate_simple_pattern())
print(len(genarate_simple_pattern()))
print(sum(1 for x in genarate_simple_pattern() if len(x)==9))

for x in  genarate_simple_pattern():
  if len(x)==9:
    visualize_pattern(x)
```

```python
visualize_pattern([1, 2, 3, 4,5,6,7,8,9])
possible_directions_simplified = {1:[2,4,5], 2:[1,3,4,5,6],3:[2,5,6],4:[1,2,5,7,8],5:[1,2,3,4,6,7,8,9],6:[2,3,5,8,9],7:[4,5,8],8:[7,4,5,6,9],9:[8,5,6]}
possible_directions_full = {1:[2,4,5,6,8], 2:[1,3,4,5,6,7,9],3:[2,5,6,4,8],4:[1,2,5,7,8,3,9],5:[1,2,3,4,6,7,8,9],6:[2,3,5,8,9,1,7],7:[4,5,8,2,6],8:[7,4,5,6,9,1,3],9:[8,5,6,4,2]}

def get_list(last_number, used):
  possible_list = possible_directions_full[last_number][:]
  possible_unlocks = {
      1:[[2,3],[5,9],[4,7]],
      2:[[5,8]],
      3:[[2,1],[6,9],[5,7]],
      4:[[5,6]],
      5:[],
      6:[[5,4]],
      7:[[4,1],[8,9],[5,3]],
      8:[[5,2]],
      9:[[5,1],[6,3],[8,7]]
      }
  for x in possible_unlocks[last_number]:
    if x[0] in used and x[-1] not in used:
      possible_list.append(x[-1])
  return [x for x in possible_list if x not in used]

def genarate_simple_pattern(start=[1],used=[]):
  used = start
  patterns = []
  patterns.append(start)
  for x in get_list(start[-1], start):
    if x not in used:
      patterns.extend(genarate_simple_pattern(start+[x]))
  return patterns
#print(genarate_simple_pattern())
print(len(genarate_simple_pattern()))
print(sum(1 for x in genarate_simple_pattern() if len(x)>=4))
for y in range(10):
  print(y, " lenght patterns: ", sum(1 for x in genarate_simple_pattern() if len(x)==y))
kaka=0
for x in  genarate_simple_pattern():
  if len(x)<4:
    kaka+=1
    print(kaka)
    visualize_pattern(x)
```
